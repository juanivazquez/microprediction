```
from microprediction import MicroReader
mr = MicroReader()
y = mr.get_lagged_values(name='hospital-er-wait-minutes-piedmont_newnan.json')
plt.plot(y)
```

# Demonstration of Stumpy on Live Data
Follows https://towardsdatascience.com/stumpy-basics-21844a2d2d92


```

m = 100
y_motifs = stump(y, m)
fig, axs = plt.subplots(2, sharex=True, gridspec_kw={'hspace': 0})
plt.suptitle('Motif (Pattern) Discovery', fontsize='30')
axs[0].plot(y)
axs[0].set_ylabel('Steam Flow', fontsize='20')
rect = Rectangle((643, 0), m, 40, facecolor='lightgrey')
axs[0].add_patch(rect)
rect = Rectangle((8724, 0), m, 40, facecolor='lightgrey')
axs[0].add_patch(rect)
axs[1].set_xlabel('Time', fontsize ='20')
axs[1].set_ylabel('Matrix Profile', fontsize='20')
axs[1].axvline(x=643, linestyle="dashed")
axs[1].axvline(x=8724, linestyle="dashed")
axs[1].plot(y_motifs[:, 0])
```
